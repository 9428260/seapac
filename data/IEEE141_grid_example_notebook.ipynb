## Gymnasium 환경 사용 예제\n\n```bash\npip install gymnasium numpy pandas\n```\n

````python
from seapac_ieee141_env import SEAPACIEEE141Env\nenv = SEAPACIEEE141Env(DATASET_PATH, bus=bus_id)\nobs, info = env.reset()\nprint('initial obs:', obs)\nprint('info:', info)\n
````

````python
total_reward = 0.0\nfor step in range(10):\n    action = env.action_space.sample()\n    obs, reward, terminated, truncated, info = env.step(action)\n    total_reward += reward\n    print(step, 'reward=', round(reward, 4), 'terminated=', terminated, 'info=', info)\n    if terminated or truncated:\n        break\nprint('total_reward=', round(total_reward, 4))\n
````
